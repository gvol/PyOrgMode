```python
# TODO: add file name, ¿line number?
def add_node(df,n,parent=None,output=None):
    if output == None:
        output = dict()

    # Handle the different types of nodes
    if type(n) == str:
        t = output.pop('text',None)
        if t == None:
            output['text'] = n
        else:
            output['text'] = t + ' ' + n
    
    elif type(n) == list:
        # print 'handling list'
        for n2 in n:
            df = add_node(df,n2,n,output)

    # Handle anything unknown
    elif not 'TYPE' in dir(n):
        print "Something with no TYPE"
        print n.__doc__
        if not type(n) in d:
            d[type(n)] = dir(d)
            a[type()] = a.__dict__
    
    # Handle a drawer
    elif n.TYPE == 'DRAWER_ELEMENT':
        #print 'handling drawer: ' + n.output()
        drawer = dict()
        for n2 in n.content:
            if type(n2) == str:
                print n.output()
            assert(n2.TYPE=='DRAWER_PROPERTY')
            output[(n.name+'.'+n2.name).lower()] = n2.value
            #df = add_node(df,n2,n,drawer)

#        for k,v in drawer:
#            output[n.name+'.'+k] = v

        return df

    # This actually shouldn't be called
    elif n.TYPE == 'DRAWER_PROPERTY':
        #print 'handling propery: ' + n.output()
        output[n.name] = n.value
        
    # TODO: not sure what to do with tables
    elif n.TYPE == 'TABLE_ELEMENT':
        pass
        #print "Table: " + str(n)
        #print n.content
        
        #t = output.pop('text',None)
        #if t == None:
        #    output['text'] = n
        #else:
        #    output['text'] = t + ' ' + n

        
    elif n.TYPE == 'SCHEDULE_ELEMENT':        
        if n.type == n.CLOSED:
            output['schedule.closed'] = n.closed
        elif n.type == n.DEADLINE:
            output['schedule.deadline'] = n.deadline
        elif n.type == n.SCHEDULED:
            output['schedule.scheduled'] = n.scheduled
        else:
            pass
        #print dir(n)

    elif n.TYPE == 'NODE_ELEMENT':
        #print "Adding level:", n.level
        # TODO: this ignores anything that we got during recursion...
        output = n.__dict__.copy()
        #output.update(n.__dict__)

        #base['type'] = n.TYPE
        #base['level'] = n.level
        #base['tags'] = n.tags
        #base['heading'] = n.heading
        #base['priority'] = n.priority
        #base['heading'] = n.heading
        #base['todo'] = n.todo

        # TODO: inherit tags from parent if not already done
        # I know these are useless
        output.pop('indent')
        output.pop('parent')
        output.pop('content')
        # Add this node to the data frame
        df.append(output)
        

    else:
        print "Something with another type: " + n.TYPE
        d[n.TYPE] = dir(d)
        a[n.TYPE] = n.__dict__

    # Recurse so that we populate `output` with anything we need
    if 'content' in dir(n):
        for n2 in n.content:
            df = add_node(df,n2,n,output)

    return df
```
